```python
import pandas as pd
import json
```

```python
#used a version of host_star_df with all stars within ~3000 LYs of earth
host_star_df = pd.read_csv('host_star.csv').set_index('HostStar')
```

```python
coords_df = host_star_df[['Coordinates', 'HabitablePlanets']].dropna()
```

```python
coords = coords_df['Coordinates']#.astype('object')
coords
```

```python
X=[]
Y=[]
Z=[]
for coordstring in coords:
    coord = coordstring.split(' ')
    X.append(float(coord[0]))
    Y.append(float(coord[1]))
    Z.append(float(coord[2]))
```

```python
names = coords_df.index.tolist()
```

```python
trace01 = {
    'x':X, 'y': Y, 'z': Z,
    'mode': 'markers',
    'text': names,
    'marker': {
        'size': [3*(1+2*ii) for ii in coords_df['HabitablePlanets'].tolist()],
        'color': ['green' if ii>0 else 'red' for ii in coords_df['HabitablePlanets'].tolist()],
        'line': {
        'width': 0.1,
        'opacity': .1
        },
        'opacity': [.1+ii for ii in coords_df['HabitablePlanets'].tolist()]
    },
    'type': 'scatter3d'
}
```
