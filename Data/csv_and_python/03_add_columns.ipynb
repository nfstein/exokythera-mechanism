```python

def lumCalc(radiusSR, temp):
    AU = 149597900000 #meters
    SR = 6.95700*(10**8) #meters
    boltzmann = 5.67*(10**(-8)) # W m^-2 K^-4 
    radius = radiusSR*SR
    lum = 4 * 3.14 * (radius**2) * boltzmann * (temp**4)
    return lum
lumCalc(1,sun_temp) #came out to the real value of 3.8e26
```

```python
radii = df.HostStarRadiusSlrRad.values.tolist() #filtering this way reveals duplicates
temps = df.HostStarTempK.values.tolist()
lums = []

for index, radius in enumerate(radii):
    print(radius, ' - ', temps[index])
    try:
        lums.append(lumCalc(float(radius), float(temps[index])))
    except ValueError:
        lums.append('nan')
        

df['HostStarLuminosity'] = lums
```

### Calculating Habitable zones from Luminosity and Mass


http://www.planetarybiology.com/calculating_habitable_zone.html

```python
def Habitable_Zones(lum):
    sun_lum = lumCalc(1,sun_temp)
    Ri = ((lum/sun_lum)/1.1)**.5
    Ro = ((lum/sun_lum)/.53)**.5
    return (Ri,Ro)
```

```python
inner_habs = []
outer_habs = []
for lum in lums:
    try:
        (Ri,Ro) = Habitable_Zones(lum)
    except TypeError:
        (Ri, Ro) = ('nan','nan')
    inner_habs.append(Ri)
    outer_habs.append(Ro)
    
df['HostStarInnerHabitabilityAU'] = inner_habs
df['HostStarOuterHabitabilityAU'] = outer_habs
outer_habs
```

### Classifying Suns

```python
stellar_classes = []
stellar_colors =[]
temp = .5
def calculateBS(temp):
    if 2400 < temp and temp < 3700:
        stellar_class = 'M'
        BC = -2.0
        color = '#ffcc6f'
    elif 3700 < temp and temp < 5200:
        stellar_class = 'K'
        BC = -0.8
        color = '#ffd2a1'
    elif 5200 < temp and temp < 6000:
        stellar_class = 'G'
        BC = -.4
        color = '#fff4ea'
    elif 6000 < temp and temp < 7500:
        stellar_class = 'F'
        BC = -.15
        color = '#f8f7ff'
    elif 7500 < temp and temp < 10000:
        stellar_class = 'A'
        BC = -.3
        color = '#cad7ff'
    elif 10000 < temp and temp < 30000:
        stellar_class = 'B'
        BC = -2.0
        color = '#aabfff'
    elif 30000 < temp:
        stellar_class = 'O'
        BC = 'nan'
        color = '#9bb0ff'
    else:
        return ('nan','nan')
    return (stellar_class, color, BC)

for index,temp in enumerate(df.HostStarTempK.values.tolist()):
    #type flag lists special circumstances of sun, 0 is an 'ordinary' singular sun
    #print(df.TypeFlag.iloc[index])
    if int(df.TypeFlag.iloc[index]) == 1:
        stellar_classes.append("P-type binary")
        stellar_colors.append('nan')
    elif int(df.TypeFlag.iloc[index]) == 2:
        stellar_classes.append("S-type binary")
        stellar_colors.append('nan')
    elif int(df.TypeFlag.iloc[index]) == 3:
        stellar_classes.append("nan") #is a lone planet
        stellar_colors.append('nan')
    elif int(df.TypeFlag.iloc[index]) == 0:
        try:
            (cla, col, bc) = calculateBS(float(temp))
            stellar_classes.append(cla)
            stellar_colors.append(col)
        except ValueError:
            stellar_classes.append('nan')
            stellar_colors.append('nan')


df['HostStarClass'] = stellar_classes
df['HostStarColor'] = stellar_colors
```

## Coordinates   

A = (RA_hours * 15) + (RA_minutes * 0.25) + (RA_seconds * 0.004166)   

B = ( ABS(Dec_degrees) + (Dec_minutes / 60) + (Dec_seconds / 3600)) * SIGN(Dec_Degrees)   

C = distance in light years or parsecs   

Cartesian:   

X = (C * cos(B)) * cos(A)   

Y = (C * cos(B)) * sin(A)   

Z = C * sin(B)

```python
coords = []
for star in host_star_df.index.tolist():
    starDat = host_star_df.loc[star]
    try: 
        #print(star, starDat['Declination'])
        DEC1 = starDat['Declination'].split(' ')

        for index, num in enumerate(DEC1):
            DEC[index] = float(num)
        RA1 = starDat['RightAscension'].split(' ')
        for index, num in enumerate(RA1):
            RA[index] = float(num)

        SIGN = np.sign(DEC[0])
        if SIGN == 0: 
            SIGN = 1

        A = RA[0]*15+RA[1]*.25+RA[2]*0.004166
        B = (abs(DEC[0]) + DEC[1]/60 + DEC[2]/3600)*(SIGN)
        C = float(starDat['DistFromSunParsec']*3.261564)
        

        X = float((C * np.cos(B)) * np.cos(A))
        Y = float((C * np.cos(B)) * np.sin(A))
        Z = float(C * np.sin(B))
        if abs(X)+abs(Y)+abs(Z) < 3000:
            coord = f"{X} {Y} {Z}"
            print(coord)
        else: 
            coord = 'nan'
            #print(f"{X} {Y} {Z}")
        #print(coord)
        #print(X,Y,Z)
    except AttributeError:
        coord = 'nan'
        print(star, 'nan')
    if star == 'sun':
        coord = str('0 0 0')
    coords.append(coord)
#print(coords)
```
