```python
import pandas as pd
# Import SQL Alchemy
import sqlalchemy
from sqlalchemy.sql import func
from sqlalchemy.orm import Session

# Import and establish Base for which classes will be constructed 
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()

# Import modules to declare columns and column data types
from sqlalchemy import Column, Integer, String, Float, Text
```

```python
engine = sqlalchemy.create_engine("sqlite:///sql/exoplanetorium.sqlite")
session = Session(bind=engine)
```

```python
planet_df = pd.read_csv('csv_and_python/working_data.csv').drop('Unnamed: 0', axis = 1)
planet_df.head()
```

```python
#planet_df.columns.tolist()
```

```python
planet_columns = [
    'PlanetIdentifier',
    'RadiusJpt',
    'PlanetaryMassJpt',
    'SemiMajorAxisAU',
    'PeriodDays',
    'Eccentricity',
    'SurfaceTempK',
    'DiscoveryMethod',
    'DiscoveryYear',
    'ListsPlanetIsOn',
    'HostStar']
```

```python
planet_df = planet_df[planet_columns]
planet_df
```

```python
host_star_df = pd.read_csv('csv_and_python/host_star.csv').rename(columns = {'HostStar': 'StarIdentifier'})
host_star_df#.columns.tolist()
```

```python
for column_name in host_star_df.columns.tolist():
    if 'HostStar' in column_name:
        print(column_name)
        host_star_df.rename(columns = {column_name: column_name.replace('HostStar', '')}, inplace=True)
host_star_df.columns.tolist()
```

```python
conn = engine.connect()
```

```python
class HostStar(Base):
    __tablename__ = "hostStar"
    StarIdentifier = Column(Text, primary_key = True)
    RadiusSlrRad = Column(Float)
    MassSlrMass = Column(Float)
    Class = Column(Text)

    TempK = Column(Float)
    InnerHabitabilityAU = Column(Float)
    OuterHabitabilityAU = Column(Float)

    Metallicity = Column(Float)
    Luminosity = Column(Float)

    PlanetList = Column(Text)
    DistFromSunLY = Column(Float)
    HabitablePlanets = Column(Text)
    HabitablePlanetsList = Column(Text)
    Planets = Column(Text)
    Coordinates = Column(Text)
    Color = Column(Text)
```

```python
class Planet(Base):
    __tablename__ = "planet"
    PlanetIdentifier = Column(Text, primary_key = True)
    RadiusJpt = Column(Float)
    PlanetaryMassJpt = Column(Float)
    SemiMajorAxisAU = Column(Float)
    PeriodDays = Column(Float)
    Eccentricity = Column(Float)
    SurfaceTempK = Column(Float)
    DiscoveryMethod = Column(Text)
    DiscoveryYear = Column(Text)
    ListsPlanetIsOn = Column(Text)
    HostStar = Column(Text) #foreign_key
    
    def __repr__(self):
        return f"id={self.PlanetIdentifier}, prcp={self.HostStar}"
```

```python
Base.metadata.create_all(engine)
```

```python
star_data = host_star_df.to_dict(orient='records')
planet_data = planet_df.to_dict(orient='records')
```

```python
metadata = sqlalchemy.MetaData(bind=engine)
metadata.reflect()
```

```python
star_table = sqlalchemy.Table("hostStar", metadata, autoload = True )
planet_table = sqlalchemy.Table("planet", metadata, autoload = True )
```

```python
conn.execute(star_table.insert(), star_data)
conn.execute(planet_table.insert(), planet_data)
```
