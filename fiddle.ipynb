```python
distance_from_sun_df = host_star_df[['DistFromSunLY','HabitablePlanets']].dropna()
bar_servable_1 = pd.value_counts(pd.cut(distance_from_sun_df['DistFromSunLY'], bins = bins))
binned = pd.DataFrame(bar_servable_1,).sort_index()
binned['Order of Distance in Light Years']=labels
binned = binned.set_index('Order of Distance in Light Years')
binned.plot(kind='bar')
plt.show()
```
