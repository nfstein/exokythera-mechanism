```python
import requests
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/OpenExoplanetCatalogue/oec_tables/master/comma_separated/open_exoplanet_catalogue.txt'
response = requests.get(url)
part = response.text.split('on\r\n# \r\n')
len(part)
```

```python
column_list = pd.DataFrame.from_csv('oec.csv').columns.tolist()
column_list
```

```python
column_list.insert(0, 'PlanetIdentifier')
column_list
```

```python
compare_column_list = part[0].split('\r\n')[5:]
for index,item in enumerate(compare_column_list):
    #print(item)
    print(item.split(':')[1] + '-' + column_list[index])
```

```python

rows = part[1].split('\r\n')
```

```python
dat_list = []
rows[0].split(',')
for row in rows:
    dat = {}
    for index,item in enumerate(row.split(',')):
        try:
            dat[column_list[index]] = item
        except:
            dat[column_list[-1]] += ', ' + item
    dat_list.append(dat)
dat_list[10]
```

```python
raw_df = pd.DataFrame(dat_list).set_index('PlanetIdentifier')
raw_df.tail(20)
```

### Kepler-953 b&c have duplicates as found in the host star loop so they are cleaned below

```python
raw_df.loc['Kepler-953 c']
```

```python

#columns_to_copy = ['SemiMajorAxisAU','InclinationDeg']
new_keplerc_row = raw_df.loc['Kepler-953 c'].iloc[1] #the first is for some other planet
new_keplerc_row
```

```python
columns_to_copy = ['SemiMajorAxisAU','InclinationDeg']
new_keplerb_row = raw_df.loc['Kepler-953 b'].iloc[1] #the first is for some other planet
for column in columns_to_copy:
    new_keplerb_row[column] = raw_df.loc['Kepler-953 b'].iloc[0][column]
new_keplerb_row
```

```python
df = raw_df.drop('Kepler-953 b').drop('Kepler-953 c')
df.loc['Kepler-953 b'] = new_keplerb_row
df.loc['Kepler-953 c'] = new_keplerc_row

df = df.dropna(how='all') # get rid of bottom allnan row

df.tail()
```

### Add HostStar Column
3 basic types of planet name declarations   
1. "HostStar letter"   
2. "HostStar.num"   
3. "Planet"   
filtered below to pull host star 

```python
#df.loc['Earth']
#filtering for star column
planets = df.index.values.tolist() #filtering this way reveals duplicates

our_planets = [
    'Mercury',
    'Venus',
    'Earth',
    'Mars',
    'Jupiter',
    'Saturn',
    'Uranus',
    'Neptune',
    'Pluto'
]
sun_column = []
sun_list = []

for index,planet in enumerate(planets):
    #ensure index is right
    #print(df.loc[planet,'PlanetaryMassJpt'])
    #print(df.iloc[index, 1])
    try:
        if df.loc[planet,'TypeFlag'] == 3:
            # TypeFlag 3 means lone planet without sun, could be none but would inhibit group_by
            sun = planet
        elif planet in our_planets:
            sun = 'sun'
        elif planet[-2] == ' ':
            sun = planet[:-2]
        elif planet[-3] == '.':
            sun = planet[:-3]
        else:
            print(df.loc[planet,'TypeFlag'], ' - ', planet) #shouldn't list any 3s but does
            sun = planet
        
        sun_column.append(sun)
        if sun not in sun_list:
            sun_list.append(sun)
            
    except IndexError:
        print(df.loc[planet])
    #print(planet)

print(index)
#print(sun_column)
```

```python
df['HostStar'] = sun_column
#check that the HostStar column lines up
df.head()
```
